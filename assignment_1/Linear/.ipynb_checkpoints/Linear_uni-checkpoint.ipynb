### SPLITTING TRAIN AND TEST DATA

```python
train = df.iloc[:1800,:]
test = df.iloc[1800:,:]
```

```python
train.info()
```

```python
test.shape
```

### EXTRACTING THE INPUT AND TARGET VARIRABLES

```python
x = train['income']
y = train['Life_expectancy']
```

### BUILDING LINEAR MULTIVARIATE MODEL

```python
def modeluni(x,y,l=0.001,itr=3000):
    m=0
    n=x.shape[0]
    c=0
    errlist = []
    for i in range(itr):
        h=m*x+c
        err=y-h
        cost=1/n*np.sum(np.square(err))
        errlist.append(cost)
        md=(2/n)*np.sum((err)*(-x))
        cd=(2/n)*np.sum(err*(-1))
        m=m-l*md
        c=c-l*cd
    return m,c,errlist
```

### CALLING THE MODEL FUNCTION BY PASSING THE PARAMETERS

```python
m,c,errlist = modeluni(x,y)
```

```python
print("SLOPE : ",m)
```

```python
print("INTERCEPT : ",c)
```

### ERROR PLOT

```python
plt.title('COST vs EPOCHS')
plt.xlabel('COST')
plt.ylabel('EPOCHS')
plt.plot(errlist)
```

```python
errlist
```

### FINAL COST

```python
errlist[-1]
```

```python
tgt_test=test['Life_expectancy']
```

### PREDICTION

```python
ypred=m*test['income']+c
```

```python
ypred
```

```python
tgt_test
```

### PREDICTION vs ACTUAL DISTRIBUTION

```python
y_and_ypred = [ypred, tgt_test]
a = plt.subplot()
labels = ['prediction', 'y_actual']
colors = ['blue','red']
a.hist(y_and_ypred, bins = 10, rwidth=0.5,label=labels, color = colors)
a.legend(prop={'size': 10})
plt.title(label="Prediction vs Actual Distribution")
plt.show()
```

# ERROR METRICS

```python
mse=np.square(np.subtract(ypred,tgt_test)).mean()
mae = np.absolute(np.subtract(ypred,tgt_test)).mean()
```

```python
print("MEAN SQUARED ERROR : ",mse)
print("MEAN ABSOLUTE ERROR : ",mae)
```
