```python
train = df.iloc[:500,:]
test = df.iloc[500:,:]
tgt_train = tgt.iloc[:500,:]
tgt_test = tgt.iloc[500:,:]
```

```python
tgt_train.shape
```

```python
x_train = train[['Glucose','Insulin']]
y_train = tgt_train
x_test = test[['Glucose','Insulin']]
y_test = tgt_test
```

```python
x_train = x_train.values
y_train = y_train.values
x_test = x_test.values
y_test = y_test.values
```

```python
x_train = x_train.T
y_train = y_train.reshape(1, x_train.shape[1])
x_test = x_test.T
y_test = y_test.reshape(1, x_test.shape[1])
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
def model_log_mul(X,Y,l=0.01,itr=1000):
    n = x_train.shape[0]  #features
    m = x_train.shape[1]  #examples
    
    w = np.zeros((n,1))
#     print(w)
    b = 0
    
    cost_list = []
    
    for i in range(itr):
        z = np.dot(w.T, X) + b
        a = sigmoid(z)
        
        #cost function
        cost = -(1/m)*np.sum(Y*np.log(a)+(1-Y)*np.log(1-a))
        
        #Gradient Descent
        dw = (1/m)*np.dot(a-Y, X.T)
        db = (1/m)*np.sum(a-Y)
        
        w = w-l*dw.T
        b = b-l*db
        
        cost_list.append(cost)
        
    return w,b,cost_list        
```

```python
w,b,cost_list = model_log_mul(x_train,y_train)
```

```python
def accuracy(X, Y, W,B):
    Z = np.dot(W.T,X)+B
    A = sigmoid(Z)
    
    A = A>0.5
    
    A = np.array(A, dtype='int64')
    acc = (1-np.sum(np.absolute(A-Y))/Y.shape[1])*100
    
    print("Accuracy of the model is : ", round(acc,2), "%")
    return A
```

```python
A = accuracy(x_test, y_test, w, b)
```

```python
from sklearn.metrics import f1_score
F1_Score = f1_score(y_test, A, average = None)
F1_Score
```
