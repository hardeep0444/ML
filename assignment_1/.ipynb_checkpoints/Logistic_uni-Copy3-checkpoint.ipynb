```python
x = train.Glucose
y = target.iloc[:500]
x_test=test.Glucose
y_test=target.iloc[500:]
n_samples = x.shape[0]
# n_features = x.shape[1]
# n_samples
```

```python
def model_log_uni(x,y,l=0.01,itr=1000):
    m=0.5
    c=1
    n_samples = x.shape[0]
    errlist = []
    for i in range(itr):
        h = m*x+c
        h_sig = (1/(1+np.exp(-h)))
        
        err = h_sig-y
        
#         cost = (1/n_samples)*np.sum(np.square(err))
        cost = -(1/n_samples)* np.sum(y*np.log(h_sig) +(1-y)*np.log(1-h_sig))
    
        errlist.append(cost)
        
        dm = (2/n_samples)*(np.sum(err *x))
        dc = (2/n_samples)*(np.sum(err))
        
        m = m-l*dm
        c = c-l*dc
    return m,c,errlist,h_sig
```

```python
m,c,errlist,h_sig = model_log_uni(x,y)
```

```python
def predict(X,w=m,bias=c):
    n=X.shape[0]
    #ypred=[]
    for i in range (n):
        linear=w*X+bias
        ypred=1/(1+np.exp(-linear))
        print(ypred)
    ans=[]
    for i in ypred:
        if(i>0.5):
            ans.append(1)
        else:
            ans.append(0)
    print(ans)
    return ans
```

```python
ypred=predict(x_test)
```
